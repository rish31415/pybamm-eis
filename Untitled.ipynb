```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def D_Teo2021(c_e):
    sigma_e = (
        4.1253e-2
        + 5.007e-4 * c_e
        - 4.7212e-7 * c_e**2
        + 1.5094e-10 * c_e**3
        - 1.6018e-14 * c_e**4
    )

    return sigma_e
```

```python
def D_Nyman2008(c_e):
    sigma_e = (
        0.1297 * (c_e / 1000) ** 3 - 2.51 * (c_e / 1000) ** 1.5 + 3.329 * (c_e / 1000)
    )

    return sigma_e
```

```python
c = np.linspace(100,5000,100)
```

```python
plt.plot(c, D_Nyman2008(c), label="Nyman")
plt.plot(c, D_Teo2021(c), label="Teo")
plt.legend()
```
